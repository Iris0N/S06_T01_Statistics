```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   
import sklearn
import statistics
```

```python
#Dades sobre fets, denúncies i trencaments de condemna per violència domèstica (Mossos d'Esquadra Open Data)
#https://mossos.gencat.cat/ca/els_mossos_desquadra/indicadors_i_qualitat/dades_obertes/cataleg_dades_obertes/

crime = pd.read_csv("/Users/fiorelliris/Documents/Coding/PYTHON/Sprint6/FetsMossos.csv")
crime.head(20)
```

```python
# Numercal Variables Description

crime.describe().round()
```

```python
crime["Any"].unique()
```

```python
crime["Àmbit"].unique()
```

```python
crime_denu = crime.groupby("Any")["Denúncies"].sum()
crime_denu = crime_denu.sort_values(ascending=False)
crime_denu.head(5)
```

```python
crime_denu_ambit = crime[["Any", "Àmbit","Denúncies"]].copy()
crime_denu_ambit.head(10)
```

```python
crime_denu_ambit.groupby(["Àmbit"]).mean().round(1)
```

```python
crime_denu_ambit_menors= crime_denu_ambit[crime_denu_ambit["Àmbit"] == "Menors"]
crime_denu_ambit_menors.head(5)
```

```python
crime_denu_ambit_menors= crime_denu_ambit_menors.groupby(["Any", "Àmbit"]).sum(["Denúncies"])
crime_denu_ambit_menors.head(10)
```

```python
list(crime_denu_ambit_menors.columns)
```

```python
crime_denu_ambit_menors.describe().round()
```

```python
#Mode:value that appears most frequently in a data set.
print(crime_denu_ambit_menors.mode())
```

```python
#Median: If the data are ordered from smallest to largest, the median is the observation right in the middle.
print(crime_denu_ambit_menors.median())
```

```python
#Standard Deviation is a measure of the amount of variation or dispersion of a set of values
#standard deviation is based on the distance from the mean.
#and it has the same units of measure as the original data. If you're talking about inches, the standard deviation will be in inches.
print(crime_denu_ambit_menors.std())
```

```python
#Compute pairwise correlation of columns, excluding NA/null values. 
#statistical measure of the strength of the relationship between the relative movements of two variables.
#The values range between -1.0 and 1.0. A calculated number greater than 1.0 or less than -1.0 means that there was an error in the correlation measurement.
#Method of correlation: pearson : standard correlation coefficient

crime.corr(method='pearson', min_periods=1)
```
